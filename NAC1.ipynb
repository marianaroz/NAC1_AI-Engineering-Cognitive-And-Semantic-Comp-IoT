```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math

#importar e tratar imagem
img = cv2.imread('circulo.png')
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (10,10))
plt.imshow(imgRGB)
```

```python
#importar e tratar imagem
img = cv2.imread('circulo.png')
imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

image_lower_azul = np.array([0, 150, 100])  
image_upper_azul = np.array([250, 250, 245])
mask_hsv = cv2.inRange(imgHSV, image_lower_azul, image_upper_azul)
contour, _ = cv2.findContours(mask_hsv, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) 
mask_rgb = cv2.cvtColor(mask_hsv, cv2.COLOR_GRAY2RGB)
contour_img = mask_rgb.copy()
cv2.drawContours(contour_img, contour, -1, [255, 0, 200], 2);

plt.figure(figsize = (10,10))  
plt.imshow(contour_img);
```

```python
color = (0,0,0)
color2 = (0,255,0)

# calculo das coordenadas do centro circ1
cnt = contour[0]
M = cv2.moments(cnt)

cx1 = int(M['m10']/M['m00'])
cy1 = int(M['m01']/M['m00'])
cv2.line(contour_img,(cx1,cy1),(cx1,cy1),color,5)

# calculo das coordenadas do centro circ2
cnt2 = contour[1]
M2 = cv2.moments(cnt2)

cx2 = int(M2['m10']/M2['m00'])
cy2 = int(M2['m01']/M2['m00'])
cv2.line(contour_img,(cx2,cy2),(cx2,cy2),color,5)

# calculo das areas
area1 = cv2.contourArea(cnt)
area2 = cv2.contourArea(cnt2)

# texto 1
font1 = cv2.FONT_HERSHEY_PLAIN 
text1 = cy2 , cx2
origin1 = (180,80)
cv2.putText(contour_img, str(text1), origin1, font1,1,(200,50,0),1,cv2.LINE_AA)

# texto 2
font2 = cv2.FONT_HERSHEY_PLAIN 
text2 = cy1 , cx1
origin2 = (400,200)
cv2.putText(contour_img, str(text2), origin2, font2,1,(200,50,0),1,cv2.LINE_AA)

# area 1
font_area1 = cv2.FONT_HERSHEY_PLAIN 
text_area1 = "Area 1 = {}".format(area1)
origin_area1 = (180,100)
cv2.putText(contour_img, str(text_area1), origin_area1, font_area1,1,(200,50,0),1,cv2.LINE_AA)

# area 2
font_area2 = cv2.FONT_HERSHEY_PLAIN 
text_area2 = "Area 2 = {}".format(area2)
origin_area2 = (400,220)
cv2.putText(contour_img, str(text_area2), origin_area2, font_area2,1,(200,50,0),1,cv2.LINE_AA)

plt.figure(figsize = (10,10))  
plt.imshow(contour_img);
```
